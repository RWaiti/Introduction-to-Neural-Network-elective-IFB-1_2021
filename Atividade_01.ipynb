Importa os pacotes necessários

```python
import matplotlib.pyplot as plt
import numpy as np
from tensorflow import keras
```

carrega os dados Iris

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

Abre imagens aleatórias do conjunto de dados para visualização em um grid de tamanho (side x side).

```python
side = 5
start = np.random.random_integers(low=0, high=x_train.shape[0], size=(1,))[0]
fig, ax = plt.subplots(side, side)
for a in range(side):
    for b in range(side):
        ax[a, b].axes.xaxis.set_visible(False)
        ax[a, b].axes.yaxis.set_visible(False)
        ax[a, b].imshow(x_train[2 * a + b + start, :, :], cmap='gray')
plt.show()
```

Normaliza os dados para treinamento e validação. Essa célula deve ser preenchida.

```python
x_train_norm = keras.utils.normalize(x_train)# insira aqui seu código
x_test_norm = keras.utils.normalize(x_test)# insira aqui seu código
```

Transforma os rótulos das imagens no formato one-hot

```python
y_train = keras.utils.to_categorical(y_train)
y_test = keras.utils.to_categorical(y_test)
```

Cria o modelo baseado numa rede neural perceptron. A rede deve possuir uma camada de entrada contendo 784 unidades (todos os valores de uma imagem de 28 x 28 pixels), uma camada oculta com o número de neurônios definido pelo estudante, e uma camada de saída com 10 neurônios.

A função de ativação da última camada deve ser escolhida para que suas saídas sejam valores entre 0 e 1.

O modelo deve ser salvo em uma variável chamada 'model'

```python
# insira aqui seu código
model = keras.Sequential()
model.add(keras.layers.Flatten(input_shape=(28,28)))
model.add(keras.layers.Dense(392,activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))
model.summary()
```

A próxima célula compila o modelo. O estudante deve escolher a função custo e o algoritmo de optimização.

```python
model.compile(loss= "categorical_crossentropy", # insira aqui seu código,
             optimizer = "adam", # insira aqui seu código,
             metrics=['accuracy'])
```

A célula seguinte treina a rede. Insira o número de épocas necessárias para o treinamento efetivo da rede. A acurácia de validação deve ficar acima de 90%.

```python
history = model.fit(x_train_norm, y_train, validation_data=(x_test_norm, y_test), epochs=10)
```
